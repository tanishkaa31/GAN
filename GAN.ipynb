```python
from tensorflow.keras.layers import Conv2D, Activation, BatchNormalization, MaxPool2D, Flatten, Dense, Conv2DTranspose, LeakyReLU, Dropout, Reshape
from tensorflow.keras import Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
```

```python
def sample_dataset(batch_size):

  random_data = []

  for _ in range(batch_size):
    name = random.choice(os.listdir('/content/train/afhq/train/wild'))
    random_img = ((show_image(f'/content/train/afhq/train/wild/{name}')))
    random_img = resize(random_img, (64,64)).astype(float32)
   # print(random_img.shape)
    random_data.append(random_img)
  return np.array(random_data)

plt.imshow(sample_dataset(5)[0])
```

## Building the Models

image.png

```python
def generator(noise_shape):
  def block(x, filters):
    x = Conv2DTranspose(filters = filters, kernel_size = (5,5), strides = (2,2), padding = "same", activation = "selu")(x)
    x = BatchNormalization()(x)
    return x

  input = Input(shape = (noise_shape,))
  x = Dense(units = 8*8*128)(input)
  x = Reshape((8,8,128))(x)
  x = block(x, 128)
  x = block(x,64)
 # x = block(x, 32)
  x = Conv2DTranspose(filters = 3, kernel_size=(4,4), strides= (2,2), padding="same", activation = 'tanh')(x)

  model = Model(inputs = input, outputs = x)
 # gen_opt = Adam(learning_rate=0.0002, beta_1=0.5)
 # model.compile(loss='binary_crossentropy', optimizer=gen_opt, metrics=['accuracy'])
  return model
```

```python
noise_shape = 100
gen_model = generator(noise_shape)
```

```python
gen_model.summary()
```

```python
def discriminator(input_shape):

  def block(x, filters):
    x = Conv2D(filters = filters, kernel_size = (3,3), strides = (2,2), padding = "same")(x)
    x = LeakyReLU(alpha = 0.2)(x)
    x = Dropout(0.3)(x)
    return x

  input = Input(shape = input_shape)
  x = block(input,128)          
  x = block(x, 64)
  x = block(x, 32)
  x = Conv2D(128,(3,3),(2,2))(x)
  x = Flatten()(x)
  #dense_2 = Dense(units = 64, activation = 'relu')(x)
  final = Dense(units = 1, activation = 'sigmoid')(x) 

  dis_opt = Adam(learning_rate=0.000001, beta_1=0.5)
  model = Model(inputs = input, outputs = final)
 # model.compile(loss='binary_crossentropy', optimizer=dis_opt, metrics=['accuracy'])
  return model
```

```python
input_shape = (64,64,3)
disc_model = discriminator(input_shape)
```

```python
def gen_noise(batch_size):
  data = []
  for _ in range(batch_size):
      x = tf.random.normal(shape = [1,100])
      data.append(x)
  data = np.array(data).reshape((batch_size,100))
  return data
```

```python
#GAN model 

disc_model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics=['accuracy'])
disc_model.trainable = False

#input_shape = (128,128,3)
input = Input(shape = (100,))
x = gen_model(input)
y = disc_model(x)
gan = Model(inputs = input, outputs = y)
gan.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics=['accuracy'])
gan.summary()
```

## Training the Model

```python
from numpy import vstack 
fake_images = []
```

```python
total_steps = 3000     #4000
batch_size = 64
epochs = 1
```

```python
for step in range(total_steps):

  avg_disc_fake_loss = []
  avg_disc_real_loss = []
  avg_GAN_loss = []

  real_X = sample_dataset(batch_size)
  real_Y = np.ones((batch_size,1)) - np.random.random_sample((batch_size,1))*0.2
  fake_X = gan.layers[1].predict(gen_noise(batch_size))
  fake_Y = np.random.random_sample((batch_size,1))*0.2

  X, y = vstack((real_X, fake_X)), vstack((real_Y, fake_Y))

  gan.layers[2].trainable = True
  gan.layers[2].train_on_batch(X,y)
  disc_metrics = gan.layers[2].train_on_batch(X,y)

  print("Disc: loss: %f" %(disc_metrics[0]))

  gan.layers[2].trainable = False
  
  for _ in range(epochs):
    gan_loss = gan.train_on_batch(gen_noise(batch_size), real_Y)
    print("Gan: loss: %f" % (gan_loss[0]))

  avg_GAN_loss.append(gan_loss[0])
  if (step+1)%500 == 0:
      fake_images.append(fake_X[0])  
  print(step)
```
